```
### import packages
import pandas as pd
import numpy as np
### show all columns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```
### Import data
df_Price_Jan= pd.read_csv("https://raw.githubusercontent.com/sunshineluyao/ESG/master/Stock_Data/Price_Jan.csv", index_col='Unnamed: 0')
df_Price_Feb= pd.read_csv("https://raw.githubusercontent.com/sunshineluyao/ESG/master/Stock_Data/Price_Feb.csv", index_col='Unnamed: 0')
df_Price_Mar= pd.read_csv("https://raw.githubusercontent.com/sunshineluyao/ESG/master/Stock_Data/Price_Mar.csv", index_col='Unnamed: 0')
df_Price_Apr= pd.read_csv("https://raw.githubusercontent.com/sunshineluyao/ESG/master/Stock_Data/Price_Apr.csv", index_col='Unnamed: 0')
df_Price_May= pd.read_csv("https://raw.githubusercontent.com/sunshineluyao/ESG/master/Stock_Data/Price_May.csv", index_col='Unnamed: 0')
df_Price_Jun= pd.read_csv("https://raw.githubusercontent.com/sunshineluyao/ESG/master/Stock_Data/Price_Jun.csv", index_col='Unnamed: 0')
df_Price_Jul= pd.read_csv("https://raw.githubusercontent.com/sunshineluyao/ESG/master/Stock_Data/Price_Jul.csv", index_col='Unnamed: 0')
```

```
df_Price=df_Price_Jan.append([df_Price_Feb,df_Price_Mar,df_Price_Apr,df_Price_May,df_Price_Jun,df_Price_Jul])
df_Price.tail()
```

```
df_Price.to_csv('Price.csv')
```
